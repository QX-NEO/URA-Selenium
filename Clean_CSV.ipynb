```python
import pandas as pd
```

```python
raw = pd.read_csv("consolidated.csv")
edit = raw
raw
```

```python
def clean_FA(data):
    split_to = data.split(" to ")
    if len(split_to) == 2:
        lower = split_to[0]
        upper = split_to[1]
        
    else:
        split_inequal = data.split(">")
        if len(split_to) == 2:
            lower = split_inequal[1]
            upper = 1000000000
        else:
            split_smaller = data.split("<= ")
            if len(split_smaller) == 2: 
                lower = 0
                upper = split_smaller[1]
            else:
                split_greater = data.split(">")
                lower = split_greater[1]
                upper = 1000000000

    return [lower, upper]
```

```python
colnames  = list(raw.columns)
list_of_sqf = list(raw[colnames[6]])
```

```python
sqf_lower = []
sqf_upper = []
for i in list_of_sqf:
    try:
        sqf_lower.append(clean_FA(i)[0])
        sqf_upper.append(clean_FA(i)[1])
    except:
        print(i)
```

```python
edit["floor_area_lower"] = sqf_lower
edit["floor_area_upper"] = sqf_upper
```

```python
col_type = [4,5,8,9]
col_names = list(edit.columns)
for i in col_type:
    edit[col_names[i]] = edit[col_names[i]].astype('float64')

edit
```

```python
edit[col_names[7]] = pd.to_datetime(edit[col_names[7]], format='%b-%y')
```

```python
edit.to_csv("clean_data.csv", header = True)
```
